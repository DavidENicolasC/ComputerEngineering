```python
import networkx as net
import graphviz
import matplotlib.pyplot as plt
import csv
```

```python
#Crea un grafo dirigido (ponderado)
g = net.DiGraph()
```

```python
#r indica cual sera el nodo raiz del grafo
r = '353'

#n indica el numero de nodos que se van a dibujar para hacer mas legible el grafo en la impresion
n = 20
```

```python
#Abre el archivo
with open('CrossActividad.csv', newline='') as archivo:
    datos = csv.reader(archivo, delimiter=',', quotechar='|')

    #i recibe los datos de la primera fila, esto es, los codigos
    i = datos.__next__()
    
    #j recibe los datos de la fila con el codigo indicado para obtener las probabilidades
    j = datos.__next__()
    while j[0] != r:
        j = datos.__next__()

    #crea los nodos y los enlaces asignando como pesos las probabilidades
    for k in range(1, n):
        if j[k] != '0%':
            g.add_edge(r, i[k], weight=int(j[k][:-1]))
```

```python
#dibuja el grafo
pos=net.spring_layout(g)
net.draw(g,node_size=500, width=0.6,with_labels=True,font_weight='bold')
labels=net.get_edge_attributes(g,'weight')
net.draw_networkx_edge_labels(g,pos,edge_labels=labels)
```
